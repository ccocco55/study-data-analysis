ğŸ“Œ í†µê³„ëŠ” ì•„ì§ ë°œìƒí•˜ì§€ ì•Šì€ ì¼ì„ ì˜ˆì¸¡í•˜ê¸° ìœ„í•´ ì‚¬ìš©í•œë‹¤.
- í†µê³„í•™ì„ ê³µë¶€í•˜ëŠ” ë° ìˆì–´ í•„ìš”í•œ ê¸°ë³¸ ê°œë…ì´ê³ ,  
  ìˆ˜ëŸ‰ì ì¸ ë¹„êµë¥¼ ê¸°ì´ˆë¡œ ë§ì€ ì‚¬ì‹¤ì„ ê´€ì°°í•˜ê³  ì²˜ë¦¬í•˜ëŠ” ë°©ë²•ì„ ì—°êµ¬í•˜ëŠ” í•™ë¬¸ì´ë‹¤.

#### ë³€ëŸ‰ (Variable)
- ìë£Œì˜ ìˆ˜ì¹˜ë¥¼ ë³€ëŸ‰ì´ë¼ê³  í•˜ë©°, ì´ëŠ” ë°ì´í„°ì˜ ê°’ì„ ì˜ë¯¸í•œë‹¤.

```python
# !pip install numpy pandas
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.randint(151, 190, size=(10, 10)), \
        columns="ì„œìš¸,ê²½ê¸°,ì¸ì²œ,ê´‘ì£¼,ëŒ€êµ¬,ë¶€ì‚°,ì „ì£¼,ê°•ë¦‰,ìš¸ì‚°,ìˆ˜ì›".split(","))

display(df)
```

#### ê³„ê¸‰ (Class)
- ë³€ëŸ‰ì„ ì¼ì • ê°„ê²©ìœ¼ë¡œ ë‚˜ëˆˆ êµ¬ê°„ì„ ì˜ë¯¸í•œë‹¤.
- ë³€ëŸ‰ì˜ ìµœì†Œê°’ê³¼ ìµœëŒ€ê°’ì„ ì˜ ê³ ë ¤í•´ì„œ ê³„ê¸‰ì„ ì •í•´ì•¼í•œë‹¤.
- ì˜ˆë¥¼ ë“¤ì–´, ê³„ê¸‰ì´ (150, 160]ì¼ ê²½ìš°, 151 ~ 160ì´ ê³„ê¸‰ì— ì†í•œë‹¤. ì¦‰ ì†Œê´„í˜¸ëŠ” êµ¬ê°„ ë¯¸í¬í•¨, ëŒ€ê´„í˜¸ëŠ” êµ¬ê°„ í¬í•¨ì´ë‹¤.

```python
df_seoul = df['ì„œìš¸']
df_class = pd.cut(df_seoul, bins=[150, 160, 170, 180, 190])
df_seoul_class = pd.DataFrame({"ì„œìš¸": df_seoul, "ê³„ê¸‰": df_class})

display(df_seoul_class)
```

#### ë„ìˆ˜ (Frequency)
- ê° ê³„ê¸‰ì— ì†í•˜ëŠ” ë³€ëŸ‰ì˜ ê°œìˆ˜ë¥¼ ì˜ë¯¸í•œë‹¤.

```python
# observed: ê±¸ê³¼ê°€ 0ê°œ ì¼ë–„ í‘œì‹œ ì—¬ë¶€, False: í‘œì‹œ, True: ë¯¸í‘œì‹œ
df_seoul_class.groupby('ê³„ê¸‰', observed=False).count()
```

#### ìƒëŒ€ ë„ìˆ˜ (Relative frequency)
- ê° ê³„ê¸‰ì— ì†í•˜ëŠ” ë³€ëŸ‰ì˜ ë¹„ìœ¨ì„ ì˜ë¯¸í•œë‹¤.

```python
df_seoul_class.groupby('ê³„ê¸‰', observed=False).count().apply(lambda x: x / 10)
```

#### ë„ìˆ˜ë¶„í¬í‘œ(Frequency table)
- ì£¼ì–´ì§„ ìë£Œë¥¼ ê³„ê¸‰ë³„ë¡œ ë‚˜ëˆˆ ë’¤ ê° ê³„ê¸‰ì— ì†í•˜ëŠ” ë„ìˆ˜ ë° ìƒëŒ€ ë„ìˆ˜ë¥¼ ì¡°ì‚¬í•œ í‘œì´ë‹¤.
- êµ¬ê°„ë³„ ë¶„í¬ë¥¼ í•œ ë²ˆì— ì•Œì•„ë³´ê¸° ì¢‹ì§€ë§Œ ê³„ê¸‰ë³„ ê° ë³€ëŸ‰ì˜ ì •í™•í•œ ê°’ì´ ìƒëµë˜ì–´ ìˆë‹¤.

```python
freq = df_seoul_class.groupby('ê³„ê¸‰', observed=False).count()['ì„œìš¸']
r_freq = df_seoul_class.groupby('ê³„ê¸‰', observed=False).count().apply(lambda x: x / 10)['ì„œìš¸']

freq_df = pd.DataFrame({'ë„ìˆ˜': freq, 'ìƒëŒ€ë„ìˆ˜': r_freq})
display(freq_df)
```

```python
freq_df.reset_index(drop=False, inplace=True)
display(freq_df)
```

#### íˆìŠ¤í† ê·¸ë¨ (Histogram)
- ë„ìˆ˜ë¶„í¬í‘œë¥¼ ì‹œê°í™”í•œ ê·¸ë˜í”„ì´ë‹¤.

```python
# !pip install matplotlib
```

```python
import matplotlib.pyplot as plt

df_seoul_class['ì„œìš¸'].hist(bins=4)
```

- ë³€ëŸ‰ì˜ í•©ì„ ë³€ëŸ‰ì˜ ìˆ˜ë¡œ ë‚˜ëˆˆ ê°’ì„ ì˜ë¯¸í•œë‹¤.

```python
df.mean(axis=0).to_frame(name="í‰ê·  í‚¤")
```

```python
g_df = df['ê²½ê¸°'].to_frame()
g_df['í¸ì°¨'] = g_df['ê²½ê¸°'].apply(lambda x: x - g_df.mean())

display(g_df)
```
